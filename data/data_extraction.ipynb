```python
import pandas as pd
```

```python
df = pd.read_csv('Co2_data.csv')
```

```python
df.keys()
```

```python
df = df[['Name', 'food counter', 'Median kg CO2eq / kg produce, Litre of Bone free meat']]
```

```python
df = df.set_index('Name')

#.to_json(orient='records',lines=True)
```

```python
df.rename(columns = {'Median kg CO2eq / kg produce, Litre of Bone free meat':'CO2eqkg', 'food counter': 'category'}, inplace = True)

```

```python
df['CO2eqkg'] = df['CO2eqkg'].map(lambda x: x.replace(',','.')).astype('float')
```
